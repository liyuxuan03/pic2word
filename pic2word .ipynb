# MintPy Word Auto Report

支持两种配置模式：
- `report_quickstart.yaml` 快速配置
- `report_schema.yaml` 高级配置

成图控制（在 YAML 中配置 `figure_generation` 段）：
- `auto` 模式：自动识别形变区域并选点
- `manual` 模式：使用 YAML 中 `manual_points` 指定的坐标
- `skip` 模式：跳过成图

要生成英文报告，将 YAML 中的 `language` 设为 `en-US`。

```python
import os
from utils import build_report_and_save

# ─── 配置区 ───
USE_ADVANCED_SCHEMA = False        # False: quickstart config, True: advanced schema config
GENERATE_FIGURES = True            # False: 跳过成图（兼容旧工作流）

CONFIG_PATH = os.path.join(
    os.getcwd(),
    'report_schema.yaml' if USE_ADVANCED_SCHEMA else 'report_quickstart.yaml'
)

print('Config file:', CONFIG_PATH)
output_path, ctx = build_report_and_save(CONFIG_PATH, generate_figures=GENERATE_FIGURES)
print('Output file:', output_path)
print('Config mode:', ctx.get('config', {}).get('_config_mode', 'unknown'))
print('Language:', ctx.get('config', {}).get('project', {}).get('language', 'unknown'))
```
